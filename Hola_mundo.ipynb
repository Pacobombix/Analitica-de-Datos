```python
#Importamos las librerias: numpy y matplotlib
#Comentario de prueba
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Definimos 2 arreglos de dimensi√≥n 100
ys = 200 + np.random.randn(100)
x = np.arange (0,100,1)
```

```python
#Gaficamos los arreglos en un plano 2D
plt.plot(x, ys,"_")
plt.fill_between(x, ys,195,facecolor="g", alpha=0.6)

#Agregamos titulo a la grafica
plt.title("Sample Visualizacion")
plt.xlabel("eje de las x", size = 16,)
plt.ylabel("eje de las y", size = 16,)
plt.show()
```
